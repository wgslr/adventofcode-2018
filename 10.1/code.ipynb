```python
import scipy.optimize
import numpy as np
import pandas as pd
```

```python
points = pd.read_csv("input.csv")
```

```python
points
```

```python
def density(t):
    movedx = points['x'] + points.vx * t
    movedy = points['y'] + points.vy * t
    return abs(max(movedx) - min(movedx)) + abs(max(movedy) - min(movedy)) 
```

```python
def display(t):
    movedx = list(points['x'] + points.vx * t)
    movedy = list(points['y'] + points.vy * t)
    minx = min(movedx)
    maxx = max(movedx)
    miny = min(movedy)
    maxy = max(movedy)
    
    for y in range(miny, maxy + 1):
        for x in range(minx, maxx + 1):
            if x in movedx and y in movedy:
                print("#", end='')
            else:
                print(" ", end='')
        print()
    
```

```python
scipy.optimize.minimize(density, 100)
```
